## Задание 2.2
Какова доля совершеннолетних пассажиров (18 лет и старше)?

```python
df[df.Age >= 18]['Name'].count() * 1.0 / df['Name'].count()
```

## Задание 2.3
Замените коды пассажирских классов (поле `Pclass`) их названиями (1 = Upper, 2 = Middle, 3 = Lower). Результат выведите на экран.

```python
def price_class(price):
    return 'Upper' if price == 1 else 'Middle' if price == 2 else 'Lower'
```

```python
df['Pclass'] = df.Pclass.apply(price_class)
```
